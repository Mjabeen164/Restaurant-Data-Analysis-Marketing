## **VARIATION OF RATINGS WITH OTHER FACTORS**

### **No. of cuisines served vs Ave ratings**

```python
#Creating a cuisine count column in rest_df
rest_df['Cuisine_count'] = cuisine_df.sum(axis=1).copy()
rest_df
```

```python
rating_by_cuisine_count = rest_df.groupby('Cuisine_count')['Aggregate rating'].mean()
rating_by_cuisine_count
```

```python
cui_count = rating_by_cuisine_count.index
mean_rating =rating_by_cuisine_count.values
```

```python
rating_by_cuisine_count.sort_values(ascending=False)
```

```python
#Plotting rating_by_cuisine_count distribution

fig,ax =plt.subplots(figsize=(6,6))
ax.bar(x =cui_count, height= mean_rating);
ax.set(xlabel = 'Number of cuisines available', ylabel='Average rating', title='Variation of ratings with the increase in cuisine count');
```

### **Ave rating vs price range**

```python
rating_by_price_range = rest_df.groupby('Price range')['Aggregate rating'].mean()
rating_by_price_range
```

```python
Pri_range = rating_by_price_range.index
ave_rating =rating_by_price_range.values
```

```python
#Plotting Rating_by_Price range distribution

fig,ax =plt.subplots(figsize=(6,6))
ax.bar(x = Pri_range, height= ave_rating);
ax.set(xticks =[1,2,3,4], xlabel = 'Price range', ylabel='Average rating', title='Variation of ratings with the increase in price range');
```

### **Ave rating vs online delivery**

```python
rating_by_online_delivery = rest_df.groupby('Has Online delivery')['Aggregate rating'].mean()
rating_by_online_delivery
```

```python
Online_deli = rating_by_online_delivery.index
aver_rating =rating_by_online_delivery.values
```

```python
#Plotting Rating_by_Online_delivery distribution

fig,ax =plt.subplots(figsize=(6,6))
ax.bar(x = Online_deli, height= aver_rating);
ax.set(xticks =[0,1], xticklabels =['No','Yes'], xlabel = 'Whether there is online delivery or not', ylabel='Average rating', title='Variation of ratings with the option of online delivery');
```

### **Ave rating vs table booking**

```python
rating_by_table_booking = rest_df.groupby('Has Table booking')['Aggregate rating'].mean()
rating_by_table_booking
```

```python
tab_book = rating_by_table_booking.index
meann_rating =rating_by_table_booking.values
```

```python
#Plotting Rating_by_Online_delivery distribution

fig,ax =plt.subplots(figsize=(6,6))
ax.bar(x = tab_book, height= meann_rating);
ax.set(xticks =[0,1], xticklabels =['No','Yes'], xlabel = 'Whether there is table booking or not', ylabel='Average rating', title='Variation of ratings with the option of table booking');
```

### **Ave Cuisine_count vs Price range**

```python
cuisine_count_by_price_range = rest_df.groupby('Price range')['Cuisine_count'].mean()
cuisine_count_by_price_range
```

```python
Pric_range = cuisine_count_by_price_range.index
ave_cui_count =cuisine_count_by_price_range.values
```

```python
#Plotting Cuisine_count_by_Price range distribution

fig,ax =plt.subplots(figsize=(6,6))
ax.bar(x = Pric_range, height= ave_cui_count);
ax.set(xticks =[1,2,3,4], xlabel = 'Price range', ylabel='Average cuisine count', title='Variation of cuisine count with the increase in price range');
```
